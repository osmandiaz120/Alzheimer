```python
#definimos las condiciones a cumplir 
conditions = [
    (df_excel["average"]>=90),
    (df_excel["average"]>=80) & ((df_excel["average"]<90)),
    (df_excel["average"]>=70) & ((df_excel["average"]<80)),
    (df_excel["average"]>=60) & ((df_excel["average"]<70)),
    (df_excel["average"]>=50) & ((df_excel["average"]<60)),
    (df_excel["average"]>=90),
]
# asiganmos los posibles valores de las condicones 
values = ["A", "B", "C", "D", "E", "F"]
# agregamos una nueva columna con las condicones y valores 
df_excel ["grades"] = np.select (conditions, values)
df_excel
```
